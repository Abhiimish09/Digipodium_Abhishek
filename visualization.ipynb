```python
#visualization
import numpy as np
import plotly.express as px
```

```python
xpoints=np.arange(40)
ypoints=np.random.rand(40)
px.line(x=xpoints, y=ypoints)
```

```python
x=np.random.randint(1,100,100)
y=np.random.randint(1,100,100)
z=np.random.uniform(1,100,100)
px.scatter_3d(x=x,y=y,z=z,color=z,size=z,size_max=18,template='plotly_dark')
```

```python
#in numpy data is stated as array
tips=px.data.tips()
tips
```

```python
#pie chart
count=tips['sex'].value_counts()
px.pie(count,count.index,count.values,template='plotly_dark')
```

```python
#donut chart
count=tips['sex'].value_counts()
px.pie(count,count.index,count.values,hole=0.5,template='plotly_dark')
```

```python
#relation of numerical data
px.scatter(tips,x='total_bill',y='tip',size='size',template='plotly_dark')
```

```python
px.scatter(tips,x='total_bill',y='tip',
           size = 'size',
           facet_col ='sex',
           facet_row ='smoker',template='plotly_dark')
```

```python
px.sunburst(
    tips,
    path=['day','time','sex'],
    values='total_bill',template='plotly_dark'
)
```

```python
iris=px.data.iris()
iris
```

```python
fig=px.scatter_matrix(iris,dimensions=['sepal_width','sepal_length','petal_width','petal_length'],
                                   color='species',
                                   height=800,width=900, template='plotly_dark')
fig.update_layout(
    title='Scatter Matrix of Iris Dataset',
    dragmode='select',
    hovermode='closest'
)
fig.show()


```

```python
import plotly.express as px
df = px.data.medals_long()

fig = px.scatter(df, y="nation", x="count", color="medal", symbol="medal")
fig.update_traces(marker_size=10)
fig.show()
```

```python
import plotly.express as px

df = px.data.gapminder().query("country=='Canada'")
fig = px.line(df, x="year", y="lifeExp", title='Life expectancy in Canada')
fig.show()
```

```python
import plotly.express as px

df = px.data.gapminder().query("continent=='Oceania'")
fig = px.line(df, x="year", y="lifeExp", color='country')
fig.show()
```

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame(dict(
    x = [1, 3, 2, 4],
    y = [1, 2, 3, 4]
))
fig = px.line(df, x="x", y="y", title="Unsorted Input") 
fig.show()

df = df.sort_values(by="x")
fig = px.line(df, x="x", y="y", title="Sorted Input") 
fig.show()

```

```python
import plotly.express as px

df = px.data.gapminder().query("country in ['Canada', 'Botswana']")

fig = px.line(df, x="lifeExp", y="gdpPercap", color="country", text="year")
fig.update_traces(textposition="bottom right")
fig.show()
```

```python
import plotly.express as px
df = px.data.gapminder()
fig = px.area(df, x="year", y="pop", color="continent", line_group="country")
fig.show()
```

```python
import plotly.express as px
df = px.data.medals_long()

fig = px.area(df, x="medal", y="count", color="nation",
             pattern_shape="nation", pattern_shape_sequence=[".", "x", "+"])
fig.show()
```
